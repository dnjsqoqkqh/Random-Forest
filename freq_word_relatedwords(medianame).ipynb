**train_data에 텍스트파일 할당**

```python
import codecs


def read_data(filename):
    with codecs.open(filename, encoding='utf-8', mode='r') as f:
        data = [line.split('\t') for line in f.read().splitlines()]
        data = data[1:]   # header 제외
    return data


train_data = read_data('instagram.txt')
```

```python
train_data
```

```python
from konlpy.tag import *
```

```python
Okt()
```

```python
from konlpy.tag import Twitter
tagger = Twitter()


def tokenize(doc):
    return ['/'.join(t) for t in tagger.pos(doc, norm=True, stem=True)]


train_docs = [row[0] for row in train_data]
```

```python
sentences = [tokenize(d) for d in train_docs]
```

```python
sentences
```

합치기

```python
nouns = []
for i in range(0,len(sentences)):
    nouns.append(sentences[i])
```

합치기2

```python
nouns2 = []
for i in range(0,len(nouns)):
    nouns2.extend(nouns[i])
```

```python
import warnings
warnings.simplefilter("ignore")

import konlpy
konlpy.__version__
```

**빈도수 세는 빈 딕셔너리 만들기**

```python
frequency = {}
```

```python
for word in nouns2:
    count = frequency.get(word,0)
    frequency[word] = count + 1
     
frequency_list = frequency.keys()
```

```python
frequency
```

```python
word_list = sorted(frequency.items(), key=lambda x: x[1], reverse=True)
```

**쓸데없는 형태소 지우기**

```python
word_list2 = []
for i in range(0,len(nouns)):
  if word_list[i][0].split('/')[1] not in ['Punctuation', 'Number', 'Suffix', 'Josa', 'Foreign', 'Conjunction', 'Determiner','Email','Exclamation','KoreanParticle','ScreenName','VerbPrefix']:
    word_list2.append(word_list[i])
```

컬럼 분리(단어, 형태소)

```python
word_list3 = []
for i in range(0, len(word_list2)):
  word_list3.append ((word_list2[i][0].split("/")[0],word_list2[i][0].split("/")[1],word_list2[i][1]))
```

```python
import pandas as pd
```

DataFrame

```python
word_list_df = pd.DataFrame(word_list3, columns=['word','morph','freq'], index=None)
```

파일 로컬 드라이브 다운

```python
from google.colab import files

f = word_list_df.to_csv("park_instagram_freqword.csv", encoding='UTF-8-sig')

files.download('park_instagram_freqword.csv')
```
